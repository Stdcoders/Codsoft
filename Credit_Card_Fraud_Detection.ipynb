```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
#loading the data set to pandas dataframe
credit_data = pd.read_csv('/content/creditcard.csv')
```

```python
#first 5 rows of the dataset
credit_data.head()
```

```python
credit_data.tail()
```

```python
# dataset information
credit_data.info()
```

```python
#missing values in the dataset
credit_data.isnull().sum()
```

```python
#filling the missing values with 0
credit_data = credit_data.fillna(0)
```

```python
#distribution of fraud transactions and legal transactions
credit_data['Class'].value_counts()
```

0 -> normal transactions
1 -> fraud transactions

```python
#seperating the data for analysis
legal = credit_data[credit_data.Class == 0]
fraud = credit_data[credit_data.Class == 1]
```

```python
print(legal.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
legal.Amount.describe()
```

```python
fraud.Amount.describe()
```

**Undersampling**

```python
legal_sample = legal.sample(n=134)

```

```python
#concatenating two dataframes
new_dataset = pd.concat([legal_sample, fraud], axis=0)
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
A = new_dataset.drop(columns='Class', axis=1)
B = new_dataset['Class']
print(A)
```

```python
print(B)
```

```python
#splitting into training and test data
A_train, A_test, B_train, B_test = train_test_split(A, B, test_size=0.2, stratify=B, random_state=2)
```

```python
print(A.shape,A_train.shape,A_test.shape)
```

```python
#model training
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
#training the model using training data
model.fit(A_train, B_train)
```

```python
#performance of the train data
X_train_prediction = model.predict(A_train)
training_data_accuracy = accuracy_score(X_train_prediction, B_train)
print('Accuracy on training data : ', training_data_accuracy)
```

```python
#performace of the test data
X_test_prediction = model.predict(A_test)
test_data_accuracy = accuracy_score(X_test_prediction, B_test)
print('Accuracy on test data : ', test_data_accuracy)
```
